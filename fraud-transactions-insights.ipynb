```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data_v1 = pd.read_csv("/kaggle/input/fraudulent-e-commerce-transactions/Fraudulent_E-Commerce_Transaction_Data.csv")
data_v2 = pd.read_csv("/kaggle/input/fraudulent-e-commerce-transactions/Fraudulent_E-Commerce_Transaction_Data_2.csv")
```

```python
# Visualization 1: Transaction Amount Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data_v1['Transaction Amount'], bins=30, kde=True)
plt.title('Transaction Amount Distribution')
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 2: Payment Method Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data_v1, x='Payment Method')
plt.title('Payment Method Distribution')
plt.xlabel('Payment Method')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 3: Fraudulent Transactions vs. Legitimate Transactions
plt.figure(figsize=(6, 6))
sns.countplot(data=data_v1, x='Is Fraudulent')
plt.title('Fraudulent Transactions vs. Legitimate Transactions')
plt.xlabel('Is Fraudulent')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 4: Product Category Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data_v1, x='Product Category')
plt.title('Product Category Distribution')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 5: Distribution of Transaction Hour
plt.figure(figsize=(10, 6))
sns.histplot(data_v1['Transaction Hour'], bins=24, kde=True)
plt.title('Distribution of Transaction Hour')
plt.xlabel('Transaction Hour')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 6: Relationship between Transaction Amount and Customer Age
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_v1, x='Transaction Amount', y='Customer Age', hue='Is Fraudulent')
plt.title('Transaction Amount vs. Customer Age')
plt.xlabel('Transaction Amount')
plt.ylabel('Customer Age')
plt.show()
```

```python
# Visualization 7: Relationship between Transaction Amount and Account Age Days
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_v1, x='Transaction Amount', y='Account Age Days', hue='Is Fraudulent')
plt.title('Transaction Amount vs. Account Age Days')
plt.xlabel('Transaction Amount')
plt.ylabel('Account Age Days')
plt.show()
```

```python
# Visualization 8: Transaction Amount by Product Category
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_v1, x='Product Category', y='Transaction Amount')
plt.title('Transaction Amount by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Transaction Amount')
plt.show()
```

```python
# Visualization 9: Distribution of Customer Age
plt.figure(figsize=(10, 6))
sns.histplot(data_v1['Customer Age'], bins=20, kde=True)
plt.title('Distribution of Customer Age')
plt.xlabel('Customer Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualization 10: Transaction Amount by Device Used
plt.figure(figsize=(10, 6))
sns.boxplot(data=data_v1, x='Device Used', y='Transaction Amount')
plt.title('Transaction Amount by Device Used')
plt.xlabel('Device Used')
plt.ylabel('Transaction Amount')
plt.show()
```

```python
# Visualization 11: Transaction Hour Distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=data_v1, x='Transaction Hour')
plt.title('Transaction Hour Distribution')
plt.xlabel('Transaction Hour')
plt.ylabel('Count')
plt.show()
```

```python
# Visualization 12: Relationship between Transaction Amount and Quantity
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data_v1, x='Transaction Amount', y='Quantity', hue='Is Fraudulent')
plt.title('Transaction Amount vs. Quantity')
plt.xlabel('Transaction Amount')
plt.ylabel('Quantity')
plt.show()
```
